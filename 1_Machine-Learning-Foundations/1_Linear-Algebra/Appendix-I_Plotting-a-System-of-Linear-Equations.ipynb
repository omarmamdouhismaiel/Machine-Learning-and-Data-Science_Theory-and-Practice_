# Plotting a System of Linear Equations

We will use:
1. **Visualization Library** _for plotting_: **Matplotlib**
2. **Numeric Library** _for simulation_: **NumPy**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 40, 1000) # start, finish, n points
```

Distance travelled by Belal: $d = 2.5t$

```python
d_b = 2.5 * t 
```

Distance travelled by Ahmed: $d = 3(t-5)$

```python
d_a = 3 * (t-5)
```

```python
fig, ax = plt.subplots()
plt.title('Ahmed & Belal Meeting')
plt.xlabel('time (in minutes)')
plt.ylabel('distance (in km)')
ax.set_xlim([0, 40])
ax.set_ylim([0, 100])
ax.plot(t, d_b, c='red')  # Belal Eq. Plot
ax.plot(t, d_a, c='blue') # Ahmed Eq. Plot
# linear system solution
plt.axvline(x=30, color='purple', linestyle='--')     # t = 30
_ = plt.axhline(y=75, color='purple', linestyle='--') # d = 75
```
